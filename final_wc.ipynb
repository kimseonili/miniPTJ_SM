# 기간별 워드클라우드

```python
import csv
import pandas as pd
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import re
# 파일불러오기
data_totals = pd.read_csv('C:/Users/song/Downloads/covid_pro/naverSearchedKeyword15.csv', encoding='cp949', header=None)
data_10s = pd.read_csv('C:/Users/song/Downloads/covid_pro/naverSearchedKeyword_10s.csv', encoding='cp949', header=None)
data_20s = pd.read_csv('C:/Users/song/Downloads/covid_pro/naverSearchedKeyword_20s.csv', encoding='cp949', header=None)
data_30s = pd.read_csv('C:/Users/song/Downloads/covid_pro/naverSearchedKeyword_30s.csv', encoding='cp949', header=None)
data_40s = pd.read_csv('C:/Users/song/Downloads/covid_pro/naverSearchedKeyword_40s.csv', encoding='cp949', header=None)
data_50s = pd.read_csv('C:/Users/song/Downloads/covid_pro/naverSearchedKeyword_50s.csv', encoding='cp949', header=None)

df_totals = pd.DataFrame(data_totals)
df_10s = pd.DataFrame(data_10s)
df_20s = pd.DataFrame(data_20s)
df_30s = pd.DataFrame(data_30s)
df_40s = pd.DataFrame(data_40s)
df_50s = pd.DataFrame(data_50s)

df_totals.columns=['month','date','keywords']
df_10s.columns=['month','date','keywords']
df_20s.columns=['month','date','keywords']
df_30s.columns=['month','date','keywords']
df_40s.columns=['month','date','keywords']
df_50s.columns=['month','date','keywords']


p = re.compile(r"[\[\'\'\]]")
df_totals['keywords']= [p.sub('', x) for x in df_totals['keywords'].tolist()]
df_10s['keywords']= [p.sub('', x) for x in df_10s['keywords'].tolist()]
df_20s['keywords']= [p.sub('', x) for x in df_20s['keywords'].tolist()]
df_30s['keywords']= [p.sub('', x) for x in df_30s['keywords'].tolist()]
df_40s['keywords']= [p.sub('', x) for x in df_40s['keywords'].tolist()]
df_50s['keywords']= [p.sub('', x) for x in df_50s['keywords'].tolist()]

# 시작, 끝 설정
start_m = int(input("시작 월:"))
start_d = int(input("시작 일:"))
end_m = int(input('끝 월:'))
end_d = int(input('끝 일'))


start = df_10s.loc[df_10s['month']==start_m][df_10s['date'] == start_d].index[0]
end = df_10s.loc[df_10s['month']==end_m][df_10s['date'] == end_d].index[0]


while_df_totals = df_totals['keywords'][start:end]
while_df_10s = df_10s['keywords'][start:end]
while_df_20s = df_20s['keywords'][start:end]
while_df_30s = df_30s['keywords'][start:end]
while_df_40s = df_40s['keywords'][start:end]
while_df_50s = df_50s['keywords'][start:end]

te_df_totals = []
te_df_10s = []
te_df_20s = []
te_df_30s = []
te_df_40s = []
te_df_50s = []

# 키워드 모으기
for i in range(start,end):
    te_df_totals.append(while_df_totals[i])
    te_df_10s.append(while_df_10s[i])
    te_df_20s.append(while_df_20s[i])
    te_df_30s.append(while_df_30s[i])
    te_df_40s.append(while_df_40s[i])
    te_df_50s.append(while_df_50s[i])
    
    

    
# 워드클라우드 생성
my_wc = WordCloud(font_path="C:/Users/song/Downloads/NanumFontSetup_TTF_ALL/NanumSquareRoundEB.ttf", background_color='white', width = 2000, height = 2000)
my_wc.generate(','.join(te_df_10s))
plt.figure(figsize = (10,10)) 
plt.imshow(my_wc)
plt.axis('off')
plt.show()

my_wc = WordCloud(font_path="C:/Users/song/Downloads/NanumFontSetup_TTF_ALL/NanumSquareRoundEB.ttf", background_color='white', width = 2000, height = 2000)
my_wc.generate(','.join(te_df_20s))
plt.figure(figsize = (10,10)) 
plt.imshow(my_wc)
plt.axis('off')
plt.show()

my_wc = WordCloud(font_path="C:/Users/song/Downloads/NanumFontSetup_TTF_ALL/NanumSquareRoundEB.ttf", background_color='white', width = 2000, height = 2000)
my_wc.generate(','.join(te_df_30s))
plt.figure(figsize = (10,10)) 
plt.imshow(my_wc)
plt.axis('off')
plt.show()

my_wc = WordCloud(font_path="C:/Users/song/Downloads/NanumFontSetup_TTF_ALL/NanumSquareRoundEB.ttf", background_color='white', width = 2000, height = 2000)
my_wc.generate(','.join(te_df_40s))
plt.figure(figsize = (10,10)) 
plt.imshow(my_wc)
plt.axis('off')
plt.show()

my_wc = WordCloud(font_path="C:/Users/song/Downloads/NanumFontSetup_TTF_ALL/NanumSquareRoundEB.ttf", background_color='white', width = 2000, height = 2000)
my_wc.generate(','.join(te_df_50s))
plt.figure(figsize = (10,10)) 
plt.imshow(my_wc)
plt.axis('off')
plt.show()

my_wc = WordCloud(font_path="C:/Users/song/Downloads/NanumFontSetup_TTF_ALL/NanumSquareRoundEB.ttf", background_color='white', width = 2000, height = 2000)
my_wc.generate(','.join(te_df_totals))
plt.figure(figsize = (10,10)) 
plt.imshow(my_wc)
plt.axis('off')
plt.show()
    
```

# 지역별 일별확진자 그래프


```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
from matplotlib.dates import MonthLocator, DateFormatter
# %matplotlib tk


font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

# 파일 불러오기
data = pd.read_csv('C:/Users/song/Downloads/covid_pro/KT_data_20200717/COVID_19/TimeProvince.csv', encoding='UTF-8', header=None)
data.columns = ['date','time','province','confirmed','released','deceased']
data.drop([0], inplace=True)

# 검색할 시, 도
pro = input('시.도 : ')

data_daegu = data[data['province'] == pro]

daegu = data_daegu[['confirmed','date']]

daegu_index = daegu['confirmed'].index



dailycon_daegu = []

for i in range(len(daegu_index)):
    if i == 0:
        dailycon_daegu.append(int(daegu['confirmed'][daegu_index[i]]))
    else:
        dailycon_daegu.append(int(daegu['confirmed'][daegu_index[i]])-int(daegu['confirmed'][daegu_index[i-1]]))
        
        
day_daegu = []
for i in daegu_index:
    day_daegu.append(daegu['date'][i])
    
# plt.figure(figsize=(20,30))
# plt.title('일별 대구 코로나 확진자',fontsize=10)
# plt.ylabel('확진자 수',fontsize=10)
# plt.xticks(rotation=90)

# plt.bar(day_daegu,dailycon_daegu,color='blue')
# plt.show()

plt.figure(figsize=(20,5))


ax = plt.subplot(1, 2, 1)
ax.xaxis.set_major_locator(MonthLocator())
ax.xaxis.set_major_formatter(DateFormatter('%b'))
# plt.xticks(rotation=90)
plt.plot(day_daegu, dailycon_daegu)
plt.title('일별 '+ pro + ' 코로나 확진자 수')

plt.show()
```
