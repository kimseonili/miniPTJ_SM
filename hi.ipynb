```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/babywalnut/Desktop/Gwangju_Project/policy_density.csv', encoding='cp949')
df_policy = pd.DataFrame(data)
# 그래프 그리기
plt.figure(figsize = (15,10)) 
#ㅇㅣ거는 그냥 데이터 표시하기위한거
plt.plot(df_policy.date, df_policy.total)
# 축 레이블
plt.xlabel('Date')
plt.ylabel('Policy density')
# x축 눈금설정
plt.xticks(['2020.01.01', '2020.02.01', '2020.03.01','2020.04.01','2020.05.01','2020.06.01','2020.07.01'],labels=['Jan', 'Feb', 'Mar','April','May','June','July'])
# 그래프 출력
plt.show()
```
