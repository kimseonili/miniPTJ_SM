```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

PatientInfo_kt = pd.read_csv(r'C:\Users\82106\Desktop\\\KT_data_20200717\COVID_19\PatientInfo_20200717.csv')

value_count_confirm = PatientInfo_kt['confirmed_date'].value_counts(bins=None)
sort_index_confirm = value_count_confirm.sort_index()


x = sort_index_confirm.index
y = sort_index_confirm.values

plt.figure(figsize=(20,5))
plt.title('  ',fontsize=20)
plt.xlabel('',fontsize=15)
plt.ylabel(' ',fontsize=15)
plt.xticks(rotation=90)
plt.bar(x,y,color='red')
plt.show()
```
